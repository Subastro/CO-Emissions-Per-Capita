```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Asus\Downloads\per-capita-co-emissions.csv")
df.info()
df.head(10)
```

```python
df["Annual CO₂ emissions (per capita)"].describe()
```

```python
# create a new datafram that contains only rows of "df" where the "Entity" column equals "Afghanistan".

df_afghan = df[df["Entity"]== "Afghanistan"]
df_afghan.info()
```

```python
x= df_afghan["Year"]
y=df_afghan["Annual CO₂ emissions (per capita)"]

plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Annual CO₂ emissions (per capita)")
plt.title("Annual CO₂ emissions (per capita) of Afghanistan")
plt.grid(True, linestyle="--",alpha=0.5)


plt.show()
```

```python
df["Entity"].nunique()
```

```python
df_usa = df[df["Entity"]== "United States"]

df_usa.info()
df_usa.head()
```

```python
df_uk = df[df["Entity"]=="United Kingdom"]
df_uk.head()
```

```python
df_china = df[df["Entity"]=="China"]
df_china.head()
```

```python
#Extracting "Year" for each country from their respective dataframe.

x1= df_china["Year"]
x2=df_usa["Year"]
x3=df_uk["Year"]

#Extracting "Annual CO₂ emissions (per capita)" for each country
yc= df_china["Annual CO₂ emissions (per capita)"]
z=df_usa["Annual CO₂ emissions (per capita)"]
s=df_uk["Annual CO₂ emissions (per capita)"]

#Now create a figure 
plt.figure(figsize=(16, 6))

#Plot CO2 emission for each country
plt.plot(x2,z,color="blue",label="USA")
plt.plot(x3,s,color="red", label='Uk')
plt.plot(x1,yc,color="purple",label="china")

# we set x-axis limits and ticks from 1750 to 2023 with 16-year intervals
plt.xlim(1750, 2023)
plt.xticks(np.arange(1750, 2024, 16))

#Similarly we set y-axis limits
plt.ylim(0,25)
plt.yticks(np.arange(0, 25, 3))

#label the x and y axis

plt.xlabel("Year", fontsize=12, labelpad=5)
plt.ylabel("Annual CO₂ emissions", fontsize=12, labelpad=5)
plt.title("CO-Emissions-Per-Capita", fontsize=16, fontweight='bold')


# Add grid and legend
plt.grid(True, linestyle="--", alpha=0.85)
plt.legend()
plt.show()
```
